## Importing the libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

## Dataset overlook:

```python
survey_2022 = pd.read_csv('survey_results_public_2022.csv')
```

```python
pd.set_option('display.max_columns', None)
survey_2022.head(3)
```

```python
survey_2022.columns
```

```python
countries = survey_2022['Country'].dropna()
%matplotlib inline
plt.figure(figsize=(10, 6))
sns.countplot(data=countries, x=countries, order=countries.value_counts().index)
plt.tick_params(axis='x', which='major', labelsize=4)
plt.xticks(rotation=90)
plt.show()
```

```python
def get_uniques(col):
    return set(survey_2022[col].str.split(';').dropna().explode())
```

```python
get_uniques('LanguageHaveWorkedWith')
```
